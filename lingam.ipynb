## Test Run with Makeup Reviews ONLY

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon if you haven't already
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# 1. Load your dataset
df_reviews = pd.read_csv('cleaned_makeup_reviews.csv')

# 2. Extract Headline Sentiment (Compound score ranges from -1 to +1)
# Ensure headlines are strings and handle NaNs
df_reviews['headline'] = df_reviews['headline'].fillna('').astype(str)
df_reviews['headline_sentiment'] = df_reviews['headline'].apply(lambda x: sia.polarity_scores(x)['compound'] if x else 0.0)

# 3. Calculate Headline Length (Effort metric)
df_reviews['headline_length'] = df_reviews['headline'].apply(lambda x: len(x.split()))

# 4. Map media type to an "Effort Score" (Text = 0, Image = 1, Video = 2)
media_map = {'text': 0, 'image': 1, 'video': 2}
# Fill missing types with 'text' or whatever your baseline code is
df_reviews['type'] = df_reviews['type'].fillna('text').astype(str).str.lower()
df_reviews['media_effort'] = df_reviews['type'].map(lambda x: media_map.get(x, 0))

# 5. Clean up core numeric columns
df_reviews['rating'] = pd.to_numeric(df_reviews['rating'], errors='coerce')
df_reviews['helpful_votes'] = pd.to_numeric(df_reviews['helpful_votes'], errors='coerce').fillna(0)

# Drop any row where rating is missing for the model
df_lingam = df_reviews[['rating', 'headline_sentiment', 'headline_length', 'media_effort', 'helpful_votes']].dropna()

print(df_lingam.head())
```

```python
import numpy as np
import pandas as pd
from causallearn.search.FCMBased import lingam

# 1. Define your feature columns explicitly
feature_names = ['rating', 'headline_sentiment', 'headline_length', 'media_effort', 'helpful_votes']

# 2. Convert your cleaned dataframe to a raw NumPy array
X = df_lingam[feature_names].to_numpy()

# 3. Initialize and fit DirectLiNGAM
model = lingam.DirectLiNGAM()
model.fit(X)

# 4. Print the Causal Order
print("--- CAUSAL ORDERING ---")
# This returns the order of variables from "root cause" to "final effect"
causal_order = [feature_names[i] for i in model.causal_order_]
print(" -> ".join(causal_order))

print("\n--- CAUSAL MATRIX ---")
# Convert the adjacency matrix into a readable DataFrame
adj_matrix_df = pd.DataFrame(model.adjacency_matrix_, index=feature_names, columns=feature_names)
print(adj_matrix_df)
```

## LiNGAM with All Data

```python
import pandas as pd
import numpy as np
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from causallearn.search.FCMBased import lingam

# --- STEP 1: INITIALIZE NLP TOOLS ---
nltk.download('vader_lexicon', quiet=True)
sia = SentimentIntensityAnalyzer()

# --- STEP 2: LOAD DATASETS ---
df_reviews = pd.read_csv('cleaned_makeup_reviews.csv')
df_products = pd.read_csv('cleaned_makeup_products.csv')

# --- STEP 3: FEATURE ENGINEERING ON REVIEWS ---
# Handle missing headlines and calculate sentiment compound score (-1 to +1)
df_reviews['headline'] = df_reviews['headline'].fillna('').astype(str)
df_reviews['headline_sentiment'] = df_reviews['headline'].apply(
    lambda x: sia.polarity_scores(x)['compound'] if x else 0.0
)

# Calculate Headline Length (User effort metric)
df_reviews['headline_length'] = df_reviews['headline'].apply(lambda x: len(x.split()))

# Map media type to a numeric Effort Score (Text = 0, Image = 1, Video = 2)
media_map = {'text': 0, 'image': 1, 'video': 2}
df_reviews['type'] = df_reviews['type'].fillna('text').astype(str).str.lower()
df_reviews['media_effort'] = df_reviews['type'].map(lambda x: media_map.get(x, 0))

# Create a cleanly named column for the review score
df_reviews['user_review_rating'] = pd.to_numeric(df_reviews['rating'], errors='coerce')
df_reviews['helpful_votes'] = pd.to_numeric(df_reviews['helpful_votes'], errors='coerce').fillna(0)


# --- STEP 4: CLEAN AND MERGE WITH PRODUCTS ---
# Clean product-level numeric data types
df_products['price'] = pd.to_numeric(df_products['price'], errors='coerce')
df_products['num_shades'] = pd.to_numeric(df_products['num_shades'], errors='coerce')

# Drop 'rating' from products completely if it exists to avoid clashes with the individual reviews
if 'rating' in df_products.columns:
    df_products = df_products.drop(columns=['rating'])

# Merge datasets together on product_link_id using a standard inner join
df_full = pd.merge(df_reviews, df_products, on='product_link_id', suffixes=('_rev_table', '_prod_table'))

# Explicit list of unique columns we are passing into the causal model
feature_names = ['price', 'num_shades', 'headline_sentiment', 'headline_length', 'media_effort', 'user_review_rating']

# Drop missing values across our target feature set
df_lingam_full = df_full[feature_names].dropna()

print(f"Data combined successfully! Matrix shape for LiNGAM: {df_lingam_full.shape}\n")
print("First 5 rows of input data:")
print(df_lingam_full.head())
print("-" * 50)
```

```python
# --- STEP 5: RUN LINGAM WITH REASONABLE SYSTEM BARS (PRIOR KNOWLEDGE) ---
X_full = df_lingam_full.to_numpy()
num_features = len(feature_names)

# Initialize background knowledge matrix (0 = allowed, 1 = forbidden direction)
tabu_edges = np.zeros((num_features, num_features))

# Map indices for rules: 
# 0: price, 1: num_shades, 2: headline_sentiment, 3: headline_length, 4: media_effort, 5: rating_review
# Rule: User behavior cannot physically dictate fixed product attributes (price/shades)
for user_behavior_idx in [2, 3, 4, 5]:
    tabu_edges[0, user_behavior_idx] = 1  # user behavior cannot cause product price
    tabu_edges[1, user_behavior_idx] = 1  # user behavior cannot cause number of shades

# Fit the DirectLiNGAM model
model = lingam.DirectLiNGAM(prior_knowledge=tabu_edges)
model.fit(X_full)

# --- STEP 6: OUTPUT RESULTS ---
print("--- LINGAM CAUSAL ORDERING ---")
causal_order = [feature_names[i] for i in model.causal_order_]
print(" -> ".join(causal_order))
print("\n(Variables upstream cause variables downstream)\n")

print("--- LINGAM CAUSAL ADJACENCY MATRIX ---")
# Build a readable map of causal impacts
adj_matrix_df = pd.DataFrame(model.adjacency_matrix_, index=feature_names, columns=feature_names)
print(adj_matrix_df)
```
