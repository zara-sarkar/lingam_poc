```python
df = pd.read_csv('cleaned_makeup_reviews.csv')
df.head(5)
```

```python
import numpy as np
import pandas as pd
from causal_learn.search.ConstraintBased.FCI import fci # Optional fallback
from causal_learn.search.FunctionalCausalModel.LiNGAM.DirectLiNGAM import DirectLiNGAM
from causal_learn.utils.GraphUtils import GraphUtils

# 1. Prepare your flat dataframe (numeric columns only, no NaNs)
# Columns order: [price, sentiment_score, review_length, rating]
X = df[['price', 'sentiment', 'review_length', 'rating']].to_numpy()

# 2. Define background knowledge (Optional but highly recommended)
# Forbid effects from causing causes (e.g., rating cannot cause price)
# tabu_edges[i, j] = 1 means forbidding an edge from i to j
tabu_edges = np.zeros((4, 4))
# Assuming 0:price, 3:rating -> Forbid rating directing to price
tabu_edges[3, 0] = 1 

# 3. Initialize and fit DirectLiNGAM
model = DirectLiNGAM(prior_knowledge=tabu_edges)
model.fit(X)

# 4. Extract results
print("Causal Ordering of Variables:")
print(model.causal_order_)

print("\nAdjacency Matrix (Causal Strengths):")
print(model.adjacency_matrix_)
```
